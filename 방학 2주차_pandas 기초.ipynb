### apply 연산
매 행에 대해 조건을 확인해야 하는 연산에는 apply만 사용이 가능하다.

```python
#대기업, 중견기업은 자산이 기준이어서 매출이랑은 상관없어요 정확한 기준이 아닙니다!
bonds['매출액']=bonds['규모'].apply(lambda x: '5000+' if x=='대기업' else '5000-1500' if x=='중견기업' else '120-1500' if x=='중기업' else '<=120')
bonds
```

## 누적 함수
변화율, 누적합, 누적곱 함수를 다시 확인해 봅시다. 덤으로 이동평균도!

1학기에 배운 거지만, 이번엔 Row 단위로 적용해 보겠습니다

```python
avante=sales22.loc['Avante (CN7)']
avante=avante[avante.columns[:-2]] #합계랑 최대 실적은 불필요하므로 삭제했습니다
avante.index=['내수', '수출']
avante
```

```python
avante.loc['내수'].pct_change() #pct_change: 변화율
```

```python
avante.loc['수출'].cumsum()
```

```python
avante.loc['수출'].pct_change().cumprod() #CAGR 연산이죠?
```

```python
#3개월 단위 내수 평균 판매실적: rolling으로 윈도우를 먼저 생성한 후 mean()함수로 평균값 계산
avante.loc['내수'].rolling(3, min_periods=1).mean()
```
