비트코인 데이터를 그려 봅시다.

빨간색 실선이 제가 결측치를 몇 개 만든 데이터이며, 원래의 데이터는 파란색으로 표시되어 있습니다.

결측치 보간은 **빨간 실선의 비어 있는 값을 파란색 별에 비슷하게 만드는** 과정입니다.

```python
coins1[['Bitcoin', 'Bitcoin_target', 'Bitcoin_ans']].plot(style=['ro-', 'k--', 'b*'],figsize=(10,5), legend=False)
```

```python
coins1[['Ethereum', 'Ethereum_target', 'Ethereum_ans']].plot(style=['mo-', 'k--', 'c*'],figsize=(10,5), legend=False)
```

가장 간단한 보간법은 **데이터의 평균**이나 **중앙값**으로 보간하는 방법입니다.

이 데이터는 기간이 짧기 때문에 이렇게 보간해도 큰 문제가 없지만,

좀 더 시기가 길어질 경우에는 그 평균이나 분산이 시점에 따라 달라질 수 있어서, 시계열 데이터에서는 적절하지 않은 기법입니다.

```python
#간단한 결측 방법: 데이터의 변동이 크지 않은 경우에만 사용
coins1['Bitcoin_mean']=coins1['Bitcoin'].fillna(coins1['Bitcoin'].mean())
coins1[['Bitcoin_mean', 'Bitcoin_target', 'Bitcoin_ans']].plot(style=['ro-', 'k--', 'b*'],figsize=(10,5), legend=False)
```

이동평균/이동중앙값으로 보간하면 어떨까요? 단순 평균/중앙값보다는 낫지만, 복잡한 변화를 제대로 쫓지 못하는 문제가 발생하기 때문에 시계열에서는 여전히 추천하지 않는 방법입니다.

```python
coins1['Bitcoin_ma']=coins1['Bitcoin'].fillna(coins1['Bitcoin'].rolling(7,min_periods=1,).mean())
coins1[['Bitcoin_ma', 'Bitcoin_target', 'Bitcoin_ans']].plot(style=['ro-', 'k--', 'b*'],figsize=(10,5), legend=False)
```

대부분의 시계열 데이터 보간은 `bfill`, `ffill`로 처리하는 것이 좋습니다. 이 기법들만 사용해도 대부분의 시계열 보간이 가능합니다.

`ffill`: 결측되지 않은 바로 이전 값으로 빈 값을 보간. LOCF (Last Observation Carried Forward)기법이라고도 지칭합니다.

`bfill`: 결측되지 않은 바로 다음 값으로 빈 값을 보간. NOCB (Next Observation Carried Backward)기법이라고도 지칭합니다.

ffill은 특히, 미래 값이 없어도 과거 값으로 바로 결측치를 메울 수 있기 때문에 실시간으로 계속해서 쌓이는 데이터에서 결측이 발생했을 경우 바로 활용이 가능하다는 강력한 장점을 갖습니다.

```python
coins1['Bitcoin_LOCF']=coins1['Bitcoin'].ffill()
coins1[['Bitcoin_LOCF', 'Bitcoin_target', 'Bitcoin_ans']].plot(style=['ro-', 'k--', 'b*'],figsize=(10,5), legend=False)
#참조할 과거 값이 없는 처음 행들은 보간되지 않습니다
```

```python
coins1['Bitcoin_NOCB']=coins1['Bitcoin'].bfill()
coins1[['Bitcoin_NOCB', 'Bitcoin_target', 'Bitcoin_ans']].plot(style=['ro-', 'k--', 'b*'],figsize=(10,5), legend=False)
```

```python
#이더리움 NOCB
coins1['Ethereum_NOCB']=coins1['Ethereum'].bfill()
coins1[['Ethereum_NOCB', 'Ethereum_target', 'Ethereum_ans']].plot(style=['mo-', 'k--', 'c*'],figsize=(10,5), legend=False)
#참조할 미래 값이 없는 마지막 행들은 보간되지 않습니다
```

그 외에, 데이터의 다른 부분들을 활용한 근사 함수를 이용하여 결측치를 보간하는 interpolate()함수가 있습니다.

선형으로 움직이는 데이터에는 잘 맞지만, 불규칙한 패턴을 갖는 데이터에는 잘 맞지 않습니다.

또 미래 값이 필요하기 때문에 실시간 데이터에는 사용이 불가하다는 단점이 있습니다.

```python
coins1['Bitcoin_linear']=coins1['Bitcoin'].interpolate(method='linear') #선형보간
coins1[['Bitcoin_linear', 'Bitcoin_target', 'Bitcoin_ans']].plot(style=['ro-', 'k--', 'b*'],figsize=(10,5), legend=False)
```

method 인자는 그 외에도,
* `quadratic` 2차식 형태 비선형 모델
* `cubic` 3차식 형태 비선형 모델
* `polynomical` 4차 이상의 고차식 형태 비선형 모델 (order 필요)
* `spline` 보간할 구간을 다시 쪼개어 부분부분 분할 (order 필요)

등 여러 가지가 있으며, 간혹 데이터의 추세를 중요하게 따르는 경우 이러한 기법들을 시도할 필요가 있습니다.

* 데이터의 정확한 구성/분포/추세를 알지 못하는 상황에서 지나치게 복잡한 보간을 수행하는 것은 통계적 왜곡 혹은 예측모델의 부정확성을 일으키기 때문에,
* quadratic, spline 외의 보간법을 실제로 사용할 일은 드물다고 보시면 됩니다.

또한 결측치가 더 많거나 보다 정밀한 예측이 필요한 경우에는 인공신경망을 만들어서 데이터에 적합한 결측치를 만드는 경우도 있습니다.

## groupby
데이터를 특정 열/인덱스를 기준으로 집계합니다.

```python
coins=pd.read_csv("/Users/yihoon-j/Documents/fr/2학기 교안 작업/Coin.csv", thousands=",",index_col=0)
```

집계를 위해서는 집계가 가능하도록 공통되는 값이 존재하는 열/인덱스가 필요하기 때문에, 먼저 인덱스를 만들어주는 작업을 수행하겠습니다.

해당 코드는 timestamp를 다루면서 다시 설명해드리겠습니다.

```python
import datetime
coins=coins.reset_index()
coins['Month']=coins['Date'].apply(lambda x: datetime.datetime.strptime(x, "%b %d, %Y").strftime('%Y%m')).astype('int64') #각 날짜의 월 번호 만들기
coins['Week']=coins['Date'].apply(lambda x: datetime.datetime.strptime(x, "%b %d, %Y").strftime('%Y%W')).astype('int64') #각 날짜의 주 번호 만들기
coins
```

groupby()안에는 집계 대상이 되는 컬럼명을, 그 뒤에는 집계 함수를 지정합니다.

집계 함수는 진짜 정말 많은데.. 제일 많이 쓰는 건 `mean()` `median()` `min()` `max()` `std()` 정도는 다들 지난 학기에 보셨을 거고,

추가로 `ohlc()`를 한번 다뤄 볼게요

```python
coins.groupby('Month').mean()
#집계가 불가능한 Date컬럼은 출력되지 않음
```

```python
#ohlc: 시가, 고가, 저가, 종가를 한꺼번에 계산
coins_ohlc=coins.drop(['Date', 'Week'], axis=1).groupby('Month').ohlc() #ohlc는 숫자형 자료형을 기준으로만 집계가 가능합니다
coins_ohlc
```

```python
#list comprehension이라는 기법 응용한 컬럼 grade down기법인데, 복붙해서 사용합시다~
coins_ohlc.columns=['_'.join(col) for col in coins_ohlc.columns]
coins_ohlc.head()
```

집계 기준, 집계 함수 각각을 여러 개로 설정할 수도 있습니다.

```python
coins.groupby(['Month', 'Week']).agg(['mean', 'ohlc'])
```

## join
두 개의 데이터프레임을 하나로 병합합니다.



실제 데이터를 사용하는 것보다는 간단한 데이터프레임으로 직접 확인하는 게 더 이해가 잘 될 거 같아서, 데이터프레임 간단히 만들어서 확인해 볼게요.

```python
#ㄱㄱ그룹 계열사 재무제표
df1=pd.DataFrame({'PER':[5.3,11.4, 3.3], 'PBR':[1.2, 0.7, 2.1]}, index=['ㄱㄱ전자', 'ㄱㄱ중공업', 'ㄱㄱ솔루션'])
df1
```

```python
#중공업 상장사 직원 수 비교
df2=pd.DataFrame({'직원수':['5천','16천', '20천']}, index=['ㄱㄱ중공업', 'ㄴㄴ중공업', 'ㄷㄷ중공업'])
df2
```

```python
df1.join(df2,how='left') #inner(default), outer, left, default
```

- 세 개 이상의 데이터프레임을 붙일 때는 데이터프레임들을`[]`로 묶어서 합칩니다. `df1.join([df2,df3,df4], how='inner']` 이런 식으로!
- 이때 left join시 가장 왼쪽 인덱스를 기준으로 조인되며, right join은 불가합니다.
- 조인되었을 때 동일한 이름의 열이 있으면 조인이 수행되지 않습니다.
